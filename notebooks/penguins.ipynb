```julia
using Deneb
```

```julia
penguins = Data(url="https://vega.github.io/vega-datasets/data/penguins.json")
```

First bar chart

```julia
chart = penguins * Mark(:bar) * Encoding(:Species, "count()")
```

Let's make a wider chart

```julia
chart *= vlspec(width=200)
```

Colored by Island

```julia
chart *= Encoding(color=:Island)
```

Also grouped by Island

```julia
chart *= Encoding(xOffset=:Island)
```

Set the width of the bars

```julia
chart *= vlspec(width=(; step = 40))
```

## Correlating two variables

```julia
x, y = "Beak Length (mm)", "Beak Depth (mm)"
```

```julia
base_scatter_plot = penguins * Mark(:point) * Encoding(
    x=(field=x, type="quantitative"),
    y=(field=y, type="quantitative"),
)
```

By default Vega-Lite always including the zero-point in quantitative axes, which is a healthy choice. However, in this case we will turn this off.

```julia
base_scatter_plot *= Encoding(x=(; scale=(; zero=false)), y=(; scale=(; zero=false)))
```

```julia
base_scatter_plot *= Encoding(color=:Species, shape=:Species)
```

```julia
base_scatter_plot * (
    vlspec() + Mark(:line) * transform_regression(x, y, groupby=:Species)
)
```
